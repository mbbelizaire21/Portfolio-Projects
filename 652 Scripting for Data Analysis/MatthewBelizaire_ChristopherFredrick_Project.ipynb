```python
#What genre of video games are most popular globally?

#The most popular Genre is Action, followed by Sports, then Misc. 
plot1 = sns.countplot(x=df["Genre"])
plot1.set_xticklabels(plot1.get_xticklabels(), rotation=60, ha="right")
plt.show
```

```python
plot2 = df.plot.scatter(x='Year_of_Release', y='NA_Sales')

'''No correlation between year released and sales number for North America.
This can indicate that while games have gotten more popular over the years, the higher quantity of games
does not autmatically translate to the quality of them which the sales would be a clear indicator of'''
```

```python
print(df.loc[df["NA_Sales"].idxmax()])

'''Wii Sports, a game from 2006 is a clear outlier in comparison with more recent titles in North America. 
Year of release has little to do with a game's success, but innovation and quality likely do. While the previous
statement likely has some truth to it, most of the original Wii consoles shipped with the console; thus increasing
the sales numbers with each console purchase in most countries'''
```

```python
#Relatively same trend as before, but games appear to sell more overall throughout time in Japan
plot3 = sns.scatterplot(data = df, x = "Year_of_Release", y = "JP_Sales", hue = "Genre")

plot3.legend(bbox_to_anchor = (1,1), fontsize = 6)
plt.show()
```

```python
print(df.loc[df["JP_Sales"].idxmax()])
'''One of the countries where Wii Sports was not sold with the system has a different game (Pokemon) as its top sales'''
```
